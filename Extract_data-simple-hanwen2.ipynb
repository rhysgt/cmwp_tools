## Extract data from solution files

This code extracts data from CMWP .sol files

```python
directory = '/home/hanwen/CMWP-211102/hanwen/'

df = extractDataDir(directory + '*/*.sol', calcLoop=False, calcLattice=False)

df = df[df.columns.drop(list(df.filter(regex='_1')))]
```

```python
samples = pd.read_csv('/home/rhys/Documents/samples.txt', delimiter='\t')
```

```python
df['pass'] = df['filename'].map(dict(zip(samples.Sample, samples['Pass'])))
df['coldroll'] = df['filename'].map(dict(zip(samples.Sample, samples['Cold Rolling'])))
```

```python
df
```

```python
passnames = df['pass'].unique()[1:]
print(passnames)

passcolours = ['r', 'g', 'b']
```

## Plot

This plots dislocation density, <a> loop fraction, M paramater and lattice paramaters as a function of depth from sample edge

```python
fig, axes = plt.subplots(2, 2, figsize=(20,14))
fig.subplots_adjust(left=0.06, right=0.96, top=0.98, bottom=0.1, hspace=0.3, wspace = 0.35)
axes = axes.ravel()


# Common settings for all axes
for sax in axes:
    sax.tick_params(axis='both', which='major', labelsize=10)
    sax.xaxis.set_ticks_position('both')
    sax.set_xlabel('Cold Rolling [%]', fontsize=12)
    sax.set_ylabel('', fontsize=12)


for name, colour in zip(passnames, passcolours):
    
    df_s = df[df['pass'] == name]   # Filter data by pass column
    
    ########################################## dislocation density

    axes[0].errorbar(x=df_s['coldroll'], y=df_s['d'], yerr = [df_s['d_neg'],df_s['d_pos']], 
                    marker='o',c=colour,  ms=5, lw=0, elinewidth=1, capsize=5, label=name)
    axes[0].set_ylabel('Dislocation density [$10^{14} m^{-2}$]', fontsize=12)
    axes[0].set_ylim(0, 25)

    ########################################## m

    axes[1].errorbar(x=df_s['coldroll'], y=np.exp(2)*df_s['e'], 
                     yerr = np.exp(2)*np.array([df_s['e_neg'],df_s['e_pos']]), 
                     marker='o',c=colour,  ms=5, lw=0, elinewidth=1, capsize=5, label=name)

    axes[1].set_ylabel('$M$', fontsize=12)
    axes[1].set_ylim(0)

    ########################################## b
    
    axes[2].errorbar(x=df_s['coldroll'], y=df_s['b'], 
                     yerr = np.array([df_s['b_neg'],df_s['b_pos']]), 
                     marker='o',c=colour,  ms=5, lw=0, elinewidth=1, capsize=5, label=name)

    axes[2].set_ylabel('$b$', fontsize=12)
    axes[2].set_ylim(0)

    ########################################## c
    
    axes[3].errorbar(x=df_s['coldroll'], y=df_s['c'], 
                     yerr = np.array([df_s['c_neg'],df_s['c_pos']]), 
                     marker='o',c=colour,  ms=5, lw=0, elinewidth=1, capsize=5, label=name)

    axes[3].set_ylabel('$c$', fontsize=12)
    axes[3].set_ylim(0)

for sax in axes:
    sax.legend(loc=1)
    
plt.show()


    
#fig.savefig('/home/rhys/Dropbox (Research Group)/BOR60.pdf', dpi=600)
```
